```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('train_qWM28Yl.csv')
df
```

```python
df.info()
```

```python
df.drop('policy_id',axis=1,inplace=True)
df.info()
```

```python
df['make'].unique()
```

```python
df['make'].replace(1,'a',inplace=True)
df['make'].replace(2,'b',inplace=True)
df['make'].replace(3,'c',inplace=True)
df['make'].replace(4,'d',inplace=True)
df['make'].replace(5,'e',inplace=True)
df['make'].unique()
```

```python
str1 = '250Nm@2750rpm'
a=str1.split('@')
a[0].rstrip('Nm')
```

```python
def Nm(str1):
    a=str1.split('@')
    return float(a[0].rstrip('Nm'))
Nm(str1)
```

```python
def rpm(str1):
    a=str1.split('@')
    return float(a[1].rstrip('rpm'))
rpm(str1)
```

```python
df['Nm']=df['max_torque'].apply(Nm)
df.to_csv('file1.csv',index=0)
```

```python
df['rpm']=df['max_torque'].apply(rpm)
df.to_csv('file1.csv',index=0)
```

```python
df.drop('max_torque',axis=1,inplace=True)
df.info()
```

```python
str1 = '40.36bhp@6000rpm'
def bhp(str1):
    a=str1.split('@')
    return float(a[0].rstrip('bhp'))
bhp(str1)
```

```python
def prpm(str1):
    a=str1.split('@')
    return float(a[1].rstrip('rpm'))
prpm(str1)
```

```python
df['bhp']=df['max_power'].apply(bhp)
df['prpm']=df['max_power'].apply(prpm)
df.drop('max_power',axis=1,inplace=True)
df.to_csv('file1.csv',index=0)
df.info()
```

```python
for column in df.columns: 
    if df[column].dtype == 'object':
        print(column.upper(),': ',df[column].nunique(),'unique values')
        print('--------------------')
        print(df[column].value_counts())
        print('\n',column.upper(),'(normalized)')
        print('--------------------')
        print(df[column].value_counts(normalize=1))
        print('\n*************************************************\n\n')
```

```python
df.isnull().sum()
```

```python
# categorical_features_indices = np.where(X.dtypes == "object")[0]
# categorical_features_indices

# df = pd.get_dummies(df,drop_first=True)
X = df.drop('is_claim', axis=1)
categorical_features_indices = np.where(X.dtypes == "object")[0]
y = df['is_claim']
```

```python
categorical_features_indices
```

```python
# pip install catboost
```

```python
from catboost import CatBoostClassifier
clf = CatBoostClassifier(random_state=1,cat_features=categorical_features_indices,learning_rate=0.5,max_depth=8)
model=clf.fit(X,y)
```
