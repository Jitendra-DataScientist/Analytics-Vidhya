# Time Series 1

```python
import pandas as pd
```

```python
df = pd.read_csv('train_IxoE5JN.csv', parse_dates = ['datetime'])
df.head()
# data is about gas production in Australia
```

```python
df.drop('row_id',axis=1,inplace=True)
df.head()
```

```python
df.set_index(keys='datetime',drop=True,inplace=True)
df.head()
```

```python
df.plot(figsize=(15,8),grid=True);
```

```python
df=df.interpolate()
```

## Moving Average Decomposition

```python
from    statsmodels.tsa.seasonal import   seasonal_decompose
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 14, 7

decomposition = seasonal_decompose(df,model='additive')
decomposition.plot();
```

A seasonal pattern in the beginning of rediduals indicate absence of strong seasonal effect in the beginning of original series.
